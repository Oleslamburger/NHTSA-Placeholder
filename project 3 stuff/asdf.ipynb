```python
#open chrome and go to NHTSA website ratings page
base_url = 'https://www.nhtsa.gov/ratings'
driver.get(base_url)
```

```python
#manufacturers list
manufacturers_list = ['Acura', 'Honda', 'BMW']
```

```python
#input manufacturer
search_box = driver.find_element(By.ID, 'ratings-search-input')
spyglass = driver.find_element(By.XPATH, '//*[@id="vehicle"]/div[2]/div[2]/div[1]/form/div/div/div/button/span')
search_box.send_keys(manufacturers_list[0])
spyglass.click()
time.sleep(1)
page_source = driver.page_source
```
